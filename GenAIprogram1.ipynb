```python
!pip install gensim
import gensim.downloader as api
word_vectors= api.load("word2vec-google-news-300")
result=word_vectors.most_similar(positive=['king','woman'],negative=['man'],topn=1)
print("king=man +woman=",result[0])
analogy_result=word_vectors.most_similar(positive=['king','woman'],negative=['man'],topn=1)
print("man:woman::king:",analogy_result[0])
similar_words=word_vectors.most_similar("car",topn=5)
print("\nTop 5 words similar to 'car':", [word for word, _ in similar_words])
```
