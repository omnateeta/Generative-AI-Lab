```python
import gensim
from gensim.models import Word2Vec
import nltk
```

```python
medical_corpus=[
    "The doctor diagnosed the patient with a server infection",
    "A hospital provides healthcare services ti patients.",
    "The nurse assisted the doctor during the surgery",
    "The patient was prescribed antibiotics to traet the infection.",
    "Doctor often recommend regular checkup to prevent illness.",
    "The surgen performed a complex operation to save the patient life.",
    "Hospitals are equipped with advanced medical technology.",
    "Vaccines helps to prevent infectious diseases.",
    "Medical research is crucial for the development of new treatments.",
    "A medical emergency requires immediate attention from healthcare professsionals."
]
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
tokenized_corpus=[word_tokenize(sentence.lower()) for sentence in medical_corpus]
```

```python
model=Word2Vec(sentences=tokenized_corpus, vector_size=100,window=5, min_count=1, workers=4)
```

```python
model.save("medical_word2vec.model")
print("model saved as 'medical_word2vec.model")
```

```python
def find_similar_words(input_word, topn=5):
  if input_word in model.wv:
    similar_words=model.wv.most_similar(input_word, topn=topn)
    return similar_words
  else:
    return f"word '{input_word}' not found in vocabulary."
```

```python
similar_words_to_doctor=find_similar_words('doctor', topn=5)
```

```python
print("\n Top semantically similar words to 'doctor':")
for word, similarity_score in similar_words_to_doctor:
  print(f"{word}: {similarity_score}")
```
